```python
def uniforme(d):
  return 0.5

def triangular(d):
  return 1 - abs(d/anchoDeBanda)

def gaussiana(d):
  return (math.exp(-(( d / anchoDeBanda )**2) / 2)) / sqrt(2 * math.pi)

def epanechnikov(d):
  return 3/4 * (1 - (d / anchoDeBanda)**2)

def varMin(d):
  return 3/8 * (3 - 5*(d / anchoDeBanda)**2)

def funcionKernel(d, tipo):
  if tipo == 1:
    return uniforme(d)
  elif tipo == 2:
    return triangular(d)
  elif tipo == 3:
    return gaussiana(d)
  elif tipo == 4:
    return epanechnikov(d)
  elif tipo == 5:
    return varMin(d)
```

```python
from math import sqrt
r = int(anchoDeBanda/resolucion)
mAcum = [[0 for i in range(Y)] for j in range(X)]
kernel = int(input('Seleccione una opcion de kernel\n 1. uniforme\n2. triangular\n3. gaussiana\n4. epanechnikov\n5. Varianza minima\n'))
for i in range(len(matriz)):
  for j in range(len(matriz[0])):
      if matriz[i][j] > 0:
        for k in range((i-r), (i+r)):
          for l in range((j-r), (j+r)):
            dist = sqrt(((i-k)**2) + ((j-l)**2)) * resolucion
            if dist <= anchoDeBanda:
              mAcum[k][l] += funcionKernel(dist, kernel) * matriz[i][j]
```
